1. Purchase Value/Session Length

```python
data.columns
```

```python
min_T = data.groupby(by=["user_id", "user_session"])["event_time"].min()
max_T = data.groupby(by=["user_id", "user_session"])["event_time"].max()
```

```python
session_length = max_T - min_T
```

```python
session_length
```

```python
avg_session_len = session_length.groupby(by="user_id").mean()
```

```python
avg_session_len = avg_session_len.apply(func=lambda x: round(x.total_seconds()/(60*60), ndigits=3))
```

```python
data["event_type"].unique()
```

```python
purchase_user_session = (
    data.assign(
        purchase_price = data["price"].where(data["event_type"] == "purchase", 0)
    )
    .groupby(["user_id", "user_session"])["purchase_price"]
    .sum()
)
```

```python
purchase_user_session = purchase_user_session.groupby(by=["user_id"]).mean()
```

```python
purchase_user_session
```

```python
avg_session_len
```

```python
purchse_session_length = purchase_user_session.divide(avg_session_len)
```

```python
brand_count["entropy"] = -brand_count["props"]*np.log2(brand_count["props"]) 
```

3. Purchase_view ratio

```python
event_counts = data.groupby(by=["user_id", "user_session", "event_type"]).size().reset_index(name="event_count")
```

```python
event_counts
```

```python
counts = event_counts.groupby(['user_id', 'user_session', 'event_type'])['event_count'].sum()
pivot_counts = counts.unstack(fill_value=0)
```

```python
counts
```

```python
pivot_counts
```

```python
pivot_counts.loc[pivot_counts["view"] == 0, "view"] = pivot_counts.loc[pivot_counts["view"] == 0, "purchase"]
```

```python
pivot_counts = pivot_counts[pivot_counts["view"] != 0]
```

```python
pivot_counts["purchase_view_ratio"] = pivot_counts["purchase"].divide(pivot_counts["view"])
```

```python
conversion_ratio = pivot_counts.groupby(by="user_id")["purchase_view_ratio"].mean()
```

```python
conversion_ratio
```

```python
conversion_ratio.describe()
```

```python
len(conversion_ratio[conversion_ratio > 0]), len(conversion_ratio[conversion_ratio == 0])
```
